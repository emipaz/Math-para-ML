Tenga en cuenta, en este caso, la función $f(\mathbf{x})$ no tiene ningún mínimo, pero a lo largo de la curva, hay dos mínimos (y dos máximos).

En una situación como ésta es donde entran en juego los multiplicadores de Lagrange. La observación es que los máximos y mínimos de la curva se encuentran donde la restricción es paralela a los contornos de la función.

Como el gradiente es perpendicular a los contornos, el gradiente de la función y el gradiente de la restricción también serán paralelos, es decir,

$\nabla f(\mathbf{x}) = \lambda \nabla g(\mathbf{x})$
∇f(x)=λ∇g(x)
Si escribimos esto en forma de componente, esto se convierte en,

$\begin{bmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \end{bmatrix} = \lambda \begin{bmatrix} \frac{\partial g}{\partial x} \\ \frac{\partial g}{\partial y} \end{bmatrix}$

Esta ecuación, junto con g(\mathbf{x}) = 0g(x)=0 es suficiente para especificar completamente el sistema. Podemos poner toda esta información en una sola ecuación vectorial,

$\nabla\mathcal{L}(x, y, \lambda) = \begin{bmatrix} \frac{\partial f}{\partial x} - \lambda \frac{\partial g}{\partial x} \\ \frac{\partial f}{\partial y} - \lambda \frac{\partial g}{\partial y} \\ -g(\mathbf{x}) \end{bmatrix} = 0$
y luego resolver esta ecuación para resolver el sistema.

Reflexionemos sobre lo que hemos hecho aquí, hemos pasado de una cuestión de encontrar un mínimo de una función 2D restringida a una curva 1D, a encontrar los ceros de una ecuación vectorial 3D.

Aunque esto pueda parecer que hemos complicado el problema, estamos perfectamente equipados para tratar este tipo de problemas; podemos utilizar los métodos de búsqueda de raíces, como el método de Newton-Raphson que hemos discutido anteriormente.

Vamos a plantear el sistema,

La función y dos de las derivadas ya están definidas. Define las otras dos sustituyendo los signos de interrogación en el siguiente código.

```python
# First we define the functions,
def f (x, y) :
    return np.exp(-(2*x*x + y*y - x*y) / 2)

def g (x, y) :
    return x*x + 3*(y+1)**2 - 1

# Next their derivatives,
def dfdx (x, y) :
    return 1/2 * (-4*x + y) * f(x, y)

def dfdy (x, y) :
    return 1/2 * (x - 2*y) * f(x, y)

def dgdx (x, y) :
    return 2*x

def dgdy (x, y) :
    return 6*(y+1)
```

A continuación vamos a definir el vector, $\nabla\mathcal{L}$ , que vamos a encontrar los ceros de; vamos a llamar a esto "DL" en el código. A continuación, podemos utilizar un método de búsqueda de raíces pre-escrito en scipy para resolver.

```python
from scipy import optimize
```

```python
def DL (xyλ) :
    [x, y, λ] = xyλ
    return np.array([
            dfdx(x, y) - λ * dgdx(x, y),
            dfdy(x, y) - λ * dgdy(x, y),
            - g(x, y)
        ])

(x0, y0, λ0) = (-1, -1, 0)
x1, y1, λ = optimize.root(DL, [x0, y0, λ0]).x
```

```python
print("x = %g" % x1)
print("y = %g" % y1)
print("λ = %g" % λ)
print("f(x, y) = %g" % f(x1, y1))
```

```python
g(x1,y1)
```

Aquí, los dos primeros elementos de la matriz son las coordenadas $x$ e $y$ que queríamos encontrar, y el último elemento es el multiplicador de Lagrange, que podemos desechar ahora que se ha utilizado.

Comprueba que (x, y) resuelve la ecuación $g(x, y) = 0$
